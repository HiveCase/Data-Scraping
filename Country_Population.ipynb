```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = urlopen("https://www.worldometers.info/world-population/population-by-country/")
```

```python
url.read()
```

```python
soup = BeautifulSoup(url.read(),'html.parser')
soup
```

```python
soup.text
```

```python
tab = soup.find_all('table')
```

```python
print(len(tab))
```

```python
# Country
tab[0].find_all('tr')[1].find_all('td')[1].text
```

```python
# Population
tab[0].find_all('tr')[1].find_all('td')[2].text
```

```python
# country and respective population
for i in range(1,234):
  print(tab[0].find_all('tr')[i].find_all('td')[1].text,tab[0].find_all('tr')[i].find_all('td')[2].text)
```

```python
import pandas as pd

# Initialize an empty list to store the extracted data
data = []

# Loop through the rows of the table and extract data
for i in range(1, 234):  # Assuming tab[0] contains the table
    row = tab[0].find_all('tr')[i]  # Get the row
    col1 = row.find_all('td')[1].text.strip()  # Extract and clean first column
    col2 = row.find_all('td')[2].text.strip()  # Extract and clean second column
    data.append([col1, col2])  # Append to list

# Convert the list to a DataFrame
df = pd.DataFrame(data, columns=['Country', 'Population'])

# Export DataFrame to CSV
df.to_csv('output.csv', index=False)

print("CSV file saved successfully!")
```
